```python
import pandas as pd
from numpy import *
from matplotlib.pyplot import * 

rc('text', usetex=True, fontsize=20)
```

```python
df_kulas = pd.read_csv('./gaussian_fit_kulas.dat', delimiter='\t')

amplitude_neg_kulas = array(df_kulas['amplitude_neg'])[0]
sigma_neg_kulas = array(df_kulas['sigma_neg'])[0]
center_neg_kulas = array(df_kulas['center_neg'])[0]
gamma_neg_kulas = array(df_kulas['gamma_neg'])[0]

amplitude_pos_kulas = array(df_kulas['amplitude_pos'])[0]
sigma_pos_kulas = array(df_kulas['sigma_pos'])[0]
center_pos_kulas = array(df_kulas['center_pos'])[0]
gamma_pos_kulas = array(df_kulas['gamma_pos'])[0]

amplitude_ratio_kulas = amplitude_neg_kulas/amplitude_pos_kulas

amplitude_neg_kulas_std = array(df_kulas['amplitude_neg_std'])[0]
sigma_neg_kulas_std = array(df_kulas['sigma_neg_std'])[0]
center_neg_kulas_std = array(df_kulas['center_neg_std'])[0]
gamma_neg_kulas_std = array(df_kulas['gamma_neg_std'])[0]

amplitude_pos_kulas_std = array(df_kulas['amplitude_pos_std'])[0]
sigma_pos_kulas_std = array(df_kulas['sigma_pos_std'])[0]
center_pos_kulas_std = array(df_kulas['center_pos_std'])[0]
gamma_pos_kulas_std = array(df_kulas['gamma_pos_std'])[0]

amplitude_ratio_kulas_std = abs(amplitude_ratio_kulas)*sqrt((amplitude_neg_kulas_std/amplitude_neg_kulas)**2+(amplitude_pos_kulas_std/amplitude_pos_kulas)**2)
```

```python
### --------------- tau10E5 --------------- ###

df = pd.read_csv('./info_tau10E5.dat', delimiter='\t')

df_vrot = df['vrot']
df_theta_lower = df['theta_lower']
thetas = unique(df_theta_lower)
thetas = append(thetas, deg2rad(90))

vrots = [50,100]

colores = ['c','b']

for t in range(len(thetas)-1):
    
    #amplitude_neg\tsigma_neg\tcenter_neg\tgamma_neg\tamplitude_pos\tsigma_pos\tcenter_pos\tgamma_pos
    
    fig_amplitude = figure(figsize=(16,8))
    
    fig_sigma_neg = figure(figsize=(16,8))
    fig_center_neg = figure(figsize=(16,8))
    fig_gamma_neg = figure(figsize=(16,8))
    
    fig_sigma_pos = figure(figsize=(16,8))
    fig_center_pos = figure(figsize=(16,8))
    fig_gamma_pos = figure(figsize=(16,8))
    
    
    ax_amplitude = fig_amplitude.add_subplot(111)
    
    ax_sigma_neg = fig_sigma_neg.add_subplot(111)
    ax_center_neg = fig_center_neg.add_subplot(111)
    ax_gamma_neg = fig_gamma_neg.add_subplot(111)
    
    ax_sigma_pos = fig_sigma_pos.add_subplot(111)
    ax_center_pos = fig_center_pos.add_subplot(111)
    ax_gamma_pos = fig_gamma_pos.add_subplot(111)
    
   
    ind_col = 0
    
    theta_lower = thetas[t]
    theta_upper = thetas[t+1]

    for vrot in vrots:

        indices_v = where((df_vrot==vrot) & (df_theta_lower==theta_lower))

        amplitude_neg = array(df['amplitude_neg'])[indices_v]
        amplitude_pos = array(df['amplitude_pos'])[indices_v]
        amplitude_ratio = amplitude_neg/amplitude_pos
        
        sigma_neg = array(df['sigma_neg'])[indices_v]
        center_neg = array(df['center_neg'])[indices_v]
        gamma_neg = array(df['gamma_neg'])[indices_v]
        
        sigma_pos = array(df['sigma_pos'])[indices_v]
        center_pos = array(df['center_pos'])[indices_v]
        gamma_pos = array(df['gamma_pos'])[indices_v]
        
        amplitude_neg_std = array(df['amplitude_neg_std'])[indices_v]
        amplitude_pos_std = array(df['amplitude_pos_std'])[indices_v]
        amplitude_ratio_std = abs(amplitude_ratio)*sqrt((amplitude_neg_std/amplitude_neg)**2+(amplitude_pos_std/amplitude_pos)**2)
        
        sigma_neg_std = array(df['sigma_neg_std'])[indices_v]
        center_neg_std = array(df['center_neg_std'])[indices_v]
        gamma_neg_std = array(df['gamma_neg_std'])[indices_v]
        
        sigma_pos_std = array(df['sigma_pos_std'])[indices_v]
        center_pos_std = array(df['center_pos_std'])[indices_v]
        gamma_pos_std = array(df['gamma_pos_std'])[indices_v]
        
        vout = array(df['vout'])[indices_v]

        vrot_lab = r'${0:.0f}$'.format(vrot)
        theta_lower_lab = r'${0:.0f}$'.format(int(rad2deg(theta_lower)))
        theta_upper_lab = r'${0:.0f}$'.format(int(rad2deg(theta_upper)))
        theta_lab = r'$\theta=$ '
        
        tit = theta_lab+theta_lower_lab+'$^\circ$ $-$ '+theta_upper_lab+'$^\circ$'
        leg = '$v_{rot}=$ '+vrot_lab
        
        ax_amplitude.plot(vout,amplitude_ratio,c=colores[ind_col])
        ax_amplitude.errorbar(vout,amplitude_ratio, yerr=amplitude_ratio_std, fmt='o',color=colores[ind_col], label=leg)
        ax_amplitude.hlines(amplitude_ratio_kulas, 0, 25, linestyle='--', color='k')
        ax_amplitude.hlines(amplitude_ratio_kulas+amplitude_ratio_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_amplitude.hlines(amplitude_ratio_kulas-amplitude_ratio_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_amplitude.set_xlim(4,22)
        ax_amplitude.set_ylim(1.6,2.3)

        ax_sigma_neg.plot(vout,sigma_neg,c=colores[ind_col])
        ax_sigma_neg.errorbar(vout,sigma_neg, yerr=sigma_neg_std, fmt='o',color=colores[ind_col], label=leg)
        ax_sigma_neg.hlines(sigma_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_sigma_neg.hlines(sigma_neg_kulas+sigma_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_sigma_neg.hlines(sigma_neg_kulas-sigma_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_sigma_neg.set_xlim(4,22)
        ax_sigma_neg.set_ylim(1,10)
        
        ax_center_neg.plot(vout,center_neg,c=colores[ind_col])
        ax_center_neg.errorbar(vout,center_neg, yerr=center_neg_std, fmt='o',color=colores[ind_col], label=leg)
        ax_center_neg.hlines(center_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_center_neg.hlines(center_neg_kulas+center_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_center_neg.hlines(center_neg_kulas-center_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_center_neg.set_xlim(4,22)
        ax_center_neg.set_ylim(-10,3)
        
        ax_gamma_neg.plot(vout,gamma_neg,c=colores[ind_col])
        ax_gamma_neg.errorbar(vout,gamma_neg, yerr=gamma_neg_std, fmt='o',color=colores[ind_col], label=leg)
        ax_gamma_neg.hlines(gamma_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_gamma_neg.hlines(gamma_neg_kulas+gamma_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_gamma_neg.hlines(gamma_neg_kulas-gamma_neg_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_gamma_neg.set_xlim(4,22)
        ax_gamma_neg.set_ylim(-7,10)

        ax_sigma_pos.plot(vout,sigma_pos,c=colores[ind_col])
        ax_sigma_pos.errorbar(vout,sigma_pos, yerr=sigma_pos_std, fmt='o',color=colores[ind_col], label=leg)
        ax_sigma_pos.hlines(sigma_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_sigma_pos.hlines(sigma_pos_kulas+sigma_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_sigma_pos.hlines(sigma_pos_kulas-sigma_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_sigma_pos.set_xlim(4,22)
        ax_sigma_pos.set_ylim(-1,20)
        
        ax_center_pos.plot(vout,center_pos,c=colores[ind_col])
        ax_center_pos.errorbar(vout,center_pos, yerr=center_pos_std, fmt='o',color=colores[ind_col], label=leg)
        ax_center_pos.hlines(center_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_center_pos.hlines(center_pos_kulas+center_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_center_pos.hlines(center_pos_kulas-center_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_center_pos.set_xlim(4,22)
        ax_center_pos.set_ylim(0,9)
        
        ax_gamma_pos.plot(vout,gamma_pos,c=colores[ind_col])
        ax_gamma_pos.errorbar(vout,gamma_pos, yerr=gamma_pos_std, fmt='o',color=colores[ind_col], label=leg)
        ax_gamma_pos.hlines(gamma_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_gamma_pos.hlines(gamma_pos_kulas+gamma_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_gamma_pos.hlines(gamma_pos_kulas-gamma_pos_kulas_std, 0, 25, linestyle='--', color='k', alpha=0.5)
        ax_gamma_pos.set_xlim(4,22)
        ax_gamma_pos.set_ylim(-10,27)

        ind_col += 1

    ax_amplitude.set_xlabel(r'$v_{out}$')
    ax_amplitude.set_ylabel(r'$A_{-}/A_{+}$')
    ax_amplitude.legend(loc='lower right')
    ax_amplitude.set_title(tit)
    fig_amplitude.savefig('tau10E5/amplitude_ratio/tau10E5_amplitude_ratio_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_sigma_neg.set_xlabel(r'$v_{out}$')
    ax_sigma_neg.set_ylabel(r'$\sigma_{-}$')
    ax_sigma_neg.legend(loc='upper left')
    ax_sigma_neg.set_title(tit)
    fig_sigma_neg.savefig('tau10E5/sigma_neg/tau10E5_sigma_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_center_neg.set_xlabel(r'$v_{out}$')
    ax_center_neg.set_ylabel(r'$c_{-}$')
    ax_center_neg.legend(loc='upper left')
    ax_center_neg.set_title(tit)
    fig_center_neg.savefig('tau10E5/center_neg/tau10E5_center_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_gamma_neg.set_xlabel(r'$v_{out}$')
    ax_gamma_neg.set_ylabel(r'$\gamma_{-}$')
    ax_gamma_neg.legend(loc='upper right')
    ax_gamma_neg.set_title(tit)
    fig_gamma_neg.savefig('tau10E5/gamma_neg/tau10E5_gamma_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_sigma_pos.set_xlabel(r'$v_{out}$')
    ax_sigma_pos.set_ylabel(r'$\sigma_{+}$')
    ax_sigma_pos.legend(loc='upper left')
    ax_sigma_pos.set_title(tit)
    fig_sigma_pos.savefig('tau10E5/sigma_pos/tau10E5_sigma_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_center_pos.set_xlabel(r'$v_{out}$')
    ax_center_pos.set_ylabel(r'$c_{+}$')
    ax_center_pos.legend(loc='upper left')
    ax_center_pos.set_title(tit)
    fig_center_pos.savefig('tau10E5/center_pos/tau10E5_center_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_gamma_pos.set_xlabel(r'$v_{out}$')
    ax_gamma_pos.set_ylabel(r'$\gamma_{+}$')
    ax_gamma_pos.legend(loc='upper left')
    ax_gamma_pos.set_title(tit)
    fig_gamma_pos.savefig('tau10E5/gamma_pos/tau10E5_gamma_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    close()
    close()
    close()
    close()
    close()
    close()
    close()
```
