```python
### --------------- tau10E5 --------------- ###

df = pd.read_csv('./info_tau10E5.dat', delimiter='\t')

df_vrot = df['vrot']
df_theta_lower = df['theta_lower']
thetas = unique(df_theta_lower)
thetas = append(thetas, deg2rad(90))

vrots = [50,100]

colores = ['c','b']

for t in range(len(thetas)-1):
    
    #amplitude_neg\tsigma_neg\tcenter_neg\tgamma_neg\tamplitude_pos\tsigma_pos\tcenter_pos\tgamma_pos
    
    fig_amplitude_neg = figure(figsize=(16,8))
    fig_sigma_neg = figure(figsize=(16,8))
    fig_center_neg = figure(figsize=(16,8))
    fig_gamma_neg = figure(figsize=(16,8))
    
    fig_amplitude_pos = figure(figsize=(16,8))
    fig_sigma_pos = figure(figsize=(16,8))
    fig_center_pos = figure(figsize=(16,8))
    fig_gamma_pos = figure(figsize=(16,8))
    
    
    ax_amplitude_neg = fig_amplitude_neg.add_subplot(111)
    ax_sigma_neg = fig_sigma_neg.add_subplot(111)
    ax_center_neg = fig_center_neg.add_subplot(111)
    ax_gamma_neg = fig_gamma_neg.add_subplot(111)
    
    ax_amplitude_pos = fig_amplitude_pos.add_subplot(111)
    ax_sigma_pos = fig_sigma_pos.add_subplot(111)
    ax_center_pos = fig_center_pos.add_subplot(111)
    ax_gamma_pos = fig_gamma_pos.add_subplot(111)
    
   
    ind_col = 0
    
    theta_lower = thetas[t]
    theta_upper = thetas[t+1]

    for vrot in vrots:

        indices_v = where((df_vrot==vrot) & (df_theta_lower==theta_lower))

        amplitude_neg = array(df['amplitude_neg'])[indices_v]
        sigma_neg = array(df['sigma_neg'])[indices_v]
        center_neg = array(df['center_neg'])[indices_v]
        gamma_neg = array(df['gamma_neg'])[indices_v]
        
        amplitude_pos = array(df['amplitude_pos'])[indices_v]
        sigma_pos = array(df['sigma_pos'])[indices_v]
        center_pos = array(df['center_pos'])[indices_v]
        gamma_pos = array(df['gamma_pos'])[indices_v]
        
        vout = array(df['vout'])[indices_v]

        vrot_lab = r'${0:.0f}$'.format(vrot)
        theta_lower_lab = r'${0:.0f}$'.format(int(rad2deg(theta_lower)))
        theta_upper_lab = r'${0:.0f}$'.format(int(rad2deg(theta_upper)))
        theta_lab = r'$\theta=$ '
        
        tit = theta_lab+theta_lower_lab+'$^\circ$ $-$ '+theta_upper_lab+'$^\circ$'
        leg = '$v_{rot}=$ '+vrot_lab
        
        ax_amplitude_neg.plot(vout,amplitude_neg,c=colores[ind_col])
        ax_amplitude_neg.scatter(vout,amplitude_neg,c=colores[ind_col],label=leg)
        ax_amplitude_neg.hlines(amplitude_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_amplitude_neg.set_xlim(4,22)
        ax_amplitude_neg.set_ylim(-50,300)

        ax_sigma_neg.plot(vout,sigma_neg,c=colores[ind_col])
        ax_sigma_neg.scatter(vout,sigma_neg,c=colores[ind_col],label=leg)
        ax_sigma_neg.hlines(sigma_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_sigma_neg.set_xlim(4,22)
        ax_sigma_neg.set_ylim(-800,50)
        
        ax_center_neg.plot(vout,center_neg,c=colores[ind_col])
        ax_center_neg.scatter(vout,center_neg,c=colores[ind_col],label=leg)
        ax_center_neg.hlines(center_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_center_neg.set_xlim(4,22)
        ax_center_neg.set_ylim(-2000,12000)
        
        ax_gamma_neg.plot(vout,gamma_neg,c=colores[ind_col])
        ax_gamma_neg.scatter(vout,gamma_neg,c=colores[ind_col],label=leg)
        ax_gamma_neg.hlines(gamma_neg_kulas, 0, 25, linestyle='--', color='k')
        ax_gamma_neg.set_xlim(4,22)
        ax_gamma_neg.set_ylim(-1000,8000)
        
        ax_amplitude_pos.plot(vout,amplitude_pos,c=colores[ind_col])
        ax_amplitude_pos.scatter(vout,amplitude_pos,c=colores[ind_col],label=leg)
        ax_amplitude_pos.hlines(amplitude_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_amplitude_pos.set_xlim(4,22)
        ax_amplitude_pos.set_ylim(-1,7)

        ax_sigma_pos.plot(vout,sigma_pos,c=colores[ind_col])
        ax_sigma_pos.scatter(vout,sigma_pos,c=colores[ind_col],label=leg)
        ax_sigma_pos.hlines(sigma_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_sigma_pos.set_xlim(4,22)
        ax_sigma_pos.set_ylim(1,7)
        
        ax_center_pos.plot(vout,center_pos,c=colores[ind_col])
        ax_center_pos.scatter(vout,center_pos,c=colores[ind_col],label=leg)
        ax_center_pos.hlines(center_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_center_pos.set_xlim(4,22)
        ax_center_pos.set_ylim(-5000,1200)
        
        ax_gamma_pos.plot(vout,gamma_pos,c=colores[ind_col])
        ax_gamma_pos.scatter(vout,gamma_pos,c=colores[ind_col],label=leg)
        ax_gamma_pos.hlines(gamma_pos_kulas, 0, 25, linestyle='--', color='k')
        ax_gamma_pos.set_xlim(4,22)
        #ax_gamma_pos.set_ylim(,)

        ind_col += 1

    ax_amplitude_neg.set_xlabel(r'$v_{out}$')
    ax_amplitude_neg.set_ylabel('$\mathrm{amplitude neg}$')
    ax_amplitude_neg.legend(loc='lower right')
    ax_amplitude_neg.set_title(tit)
    fig_amplitude_neg.savefig('tau10E5/amplitude_neg/tau10E5_amplitude_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_sigma_neg.set_xlabel(r'$v_{out}$')
    ax_sigma_neg.set_ylabel('$\mathrm{sigma neg}$')
    ax_sigma_neg.legend(loc='lower right')
    ax_sigma_neg.set_title(tit)
    fig_sigma_neg.savefig('tau10E5/sigma_neg/tau10E5_sigma_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_center_neg.set_xlabel(r'$v_{out}$')
    ax_center_neg.set_ylabel('$\mathrm{center neg}$')
    ax_center_neg.legend(loc='lower right')
    ax_center_neg.set_title(tit)
    fig_center_neg.savefig('tau10E5/center_neg/tau10E5_center_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_gamma_neg.set_xlabel(r'$v_{out}$')
    ax_gamma_neg.set_ylabel('$\mathrm{gamma neg}$')
    ax_gamma_neg.legend(loc='lower right')
    ax_gamma_neg.set_title(tit)
    fig_gamma_neg.savefig('tau10E5/gamma_neg/tau10E5_gamma_neg_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_amplitude_pos.set_xlabel(r'$v_{out}$')
    ax_amplitude_pos.set_ylabel('$\mathrm{amplitude pos}$')
    ax_amplitude_pos.legend(loc='lower right')
    ax_amplitude_pos.set_title(tit)
    fig_amplitude_pos.savefig('tau10E5/amplitude_pos/tau10E5_amplitude_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_sigma_pos.set_xlabel(r'$v_{out}$')
    ax_sigma_pos.set_ylabel('$\mathrm{sigma pos}$')
    ax_sigma_pos.legend(loc='lower right')
    ax_sigma_pos.set_title(tit)
    fig_sigma_pos.savefig('tau10E5/sigma_pos/tau10E5_sigma_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_center_pos.set_xlabel(r'$v_{out}$')
    ax_center_pos.set_ylabel('$\mathrm{center pos}$')
    ax_center_pos.legend(loc='lower right')
    ax_center_pos.set_title(tit)
    fig_center_pos.savefig('tau10E5/center_pos/tau10E5_center_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    ax_gamma_pos.set_xlabel(r'$v_{out}$')
    ax_gamma_pos.set_ylabel('$\mathrm{gamma pos}$')
    ax_gamma_pos.legend(loc='lower right')
    ax_gamma_pos.set_title(tit)
    fig_gamma_pos.savefig('tau10E5/gamma_pos/tau10E5_gamma_pos_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')
    
    close()
    close()
    close()
    close()
    close()
    close()
    close()
    close()
```

```python
### --------------- tau10E6 --------------- ###

df = pd.read_csv('./info_tau10E6.dat', delimiter='\t')

df_vrot = df['vrot']
df_theta_lower = df['theta_lower']
thetas = unique(df_theta_lower)
thetas = append(thetas, deg2rad(90))

vrots = [50,100]

colores = ['c','b']

for t in range(len(thetas)-1):
    
    fig_skw = figure(figsize=(16,8))
    ax_skw = fig_skw.add_subplot(111)

    fig_std = figure(figsize=(16,8))
    ax_std = fig_std.add_subplot(111)
    
    ind_col = 0
    
    theta_lower = thetas[t]
    theta_upper = thetas[t+1]

    for vrot in vrots:

        indices_v = where((df_vrot==vrot) & (df_theta_lower==theta_lower))

        skw = array(df['skw'])[indices_v]
        std = array(df['std'])[indices_v]
        vout = array(df['vout'])[indices_v]

        vrot_lab = r'${0:.0f}$'.format(vrot)
        theta_lower_lab = r'${0:.0f}$'.format(int(rad2deg(theta_lower)))
        theta_upper_lab = r'${0:.0f}$'.format(int(rad2deg(theta_upper)))
        theta_lab = r'$\theta=$ '
        
        tit = theta_lab+theta_lower_lab+'$^\circ$ $-$ '+theta_upper_lab+'$^\circ$'
        leg = '$v_{rot}=$ '+vrot_lab
        
        ax_skw.plot(vout,skw,c=colores[ind_col])
        ax_skw.scatter(vout,skw,color=colores[ind_col],label=leg)
        ax_skw.set_ylim(0.15,1.05)

        ax_std.plot(vout,std,c=colores[ind_col])
        ax_std.scatter(vout,std,color=colores[ind_col],label=leg)
        ax_std.set_ylim(6.5,8.5)

        ind_col += 1

    ax_skw.set_xlabel(r'$v_{out}$')
    ax_skw.set_ylabel('$\mathrm{skw}$')
    ax_skw.legend(loc='lower right')
    ax_skw.set_title(tit)
    fig_skw.savefig('tau10E6/skw/tau10E6_skw_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    ax_std.set_xlabel(r'$v_{out}$')
    ax_std.set_ylabel('$\mathrm{std}$')
    ax_std.legend(loc = 'best')
    ax_std.set_title(tit)
    fig_std.savefig('tau10E6/std/tau10E6_std_phi'+str(int(rad2deg(theta_lower)))+'-'+str(int(rad2deg(theta_upper)))+'.png', bbox_inches='tight')

    close()
    close()
```
