```python
%pylab inline

rc('text', usetex=True, fontsize=28)
```

```python
data = loadtxt('data.csv', delimiter=',')

x = data[:,0]/(1+3.50618) #redshift z = 3.50618 \pm 0.00005
y = data[:,1]
```

```python
f = figure(figsize = (16,10))
plot(x,y)
xlabel('$\mathrm{V\;(km\;s^{-1})}$')
ylabel('$\mathrm{Flux\;(10^{-17}\;erg/s/cm^{2}\;/\AA)}$')
savefig('observation_v.png')
```

```python
v_th = -12.8
```

```python
f = figure(figsize = (16,10))
plot(x/v_th,y)
xlabel('$\mathrm{x}$')
ylabel('$\mathrm{Flux\;(10^{-17}\;erg/s/cm^{2}\;/\AA)}$')
savefig('observation_x.png')
```
