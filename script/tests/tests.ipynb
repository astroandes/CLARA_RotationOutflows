```python
%pylab inline
import pandas as pd

rc('text', usetex=True, fontsize=20)
```

```python
### --------------- STATIC --------------- ###

rot = pd.read_csv('data/test_rotationstatic/TestVoutflow0Vrotation0_out.ascii', delimiter=' ')
rotout = pd.read_csv('data/test_rotationoutflowstatic/TestVoutflow0Vrotation0_out.ascii', delimiter=' ')
```

```python
rot
```

```python
rotout
```

```python
rot_xfrec = rot['x_frec']
rotout_xfrec = rotout['x_frec']
```

```python
figure(figsize=(16,10))

n,b,p = hist(rot_xfrec, color='y', alpha=0.5, label='$\mathrm{Rotation}$')
n,b,p = hist(rotout_xfrec, color='b', alpha=0.5, label='$\mathrm{Rotation+Outflow}$')

xlabel('$x$')
ylabel('$I$')
title('$\mathrm{STATIC}$')
legend()
```

```python
print len(rot_xfrec)
print len(where(rot_xfrec-rotout_xfrec == 0)[0])
```
