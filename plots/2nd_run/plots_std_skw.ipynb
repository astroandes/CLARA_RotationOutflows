```python
import pandas as pd
from numpy import *
from matplotlib.pyplot import * 

rc('text', usetex=True, fontsize=20)
```

```python
### --------------- tau10E5 --------------- ###

df = pd.read_csv('./info_tau10E5.dat', delimiter='\t')

df_vrot = df['vrot']
df_vout = df['vout']

vrots = [50,100]
vouts = [25,50,75]

colores = ['c','b','k']

for vrot in vrots:
    
    fig_skw = figure(figsize=(12,8))
    ax_skw = fig_skw.add_subplot(111)
    
    fig_std = figure(figsize=(12,8))
    ax_std = fig_std.add_subplot(111)
    
    ind_col = 0
    
    for vout in vouts:
        
        indices_v = where((df_vrot==vrot) & (df_vout==vout))
        
        skw = array(df['skw'])[indices_v]
        std = array(df['std'])[indices_v]
        thetas = array(df['theta_lower'])[indices_v]
        
        vrot_lab = r'${0:.0f}$'.format(vrot)
        vout_lab = r'${0:.0f}$'.format(vout)
        lab = '$v_{out}=$ '+vout_lab
        
        ax_skw.plot(thetas,skw,c=colores[ind_col],label=lab)
        ax_skw.set_ylim(-0.1,0.7)
        
        ax_std.plot(thetas,std,c=colores[ind_col],label=lab)
        ax_std.set_ylim(4.2,5.6)
        
        ind_col += 1
        
        
    ax_skw.set_xlabel(r'$\theta$')
    ax_skw.set_ylabel('$\mathrm{skw}$')
    ax_skw.legend()
    ax_skw.set_title('$v_{rot}=$ '+vrot_lab)
    fig_skw.savefig('tau10E5/skw/vrot'+str(vrot)+'_skw.png')
    
    ax_std.set_xlabel(r'$\theta$')
    ax_std.set_ylabel('$\mathrm{std}$')
    ax_std.legend()
    ax_std.set_title('$v_{rot}=$ '+vrot_lab)
    fig_std.savefig('tau10E5/std/vrot'+str(vrot)+'_std.png')
        
    close()
    close()
```
